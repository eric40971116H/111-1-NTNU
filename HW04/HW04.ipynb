```python
import jieba
import jieba.analyse
import pandas as pd
import numpy as np
from collections import Counter
```

```python
sf['seg'] = ""
sf.head()
```

```python
jieba.load_userdict('userdict.txt')
```

```python
jieba.analyse.set_stop_words('stopwords.txt')
index = 0
for documents in sf[0]:
    seg_list = jieba.analyse.extract_tags(documents, 100)
    #seg_list = jieba.lcut(documents)
    sf['seg'].iloc[index] = seg_list
    index += 1
```

```python
subdata_seg_list = sf['seg'].tolist()
```

```python
flat = [x for sublist in subdata_seg_list for x in sublist]
diction = Counter(flat)
```
