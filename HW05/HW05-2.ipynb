```python
import pandas as pd
from collections import Counter
orgdata = pd.read_csv("17.csv")
orgdata.head()
orgdata.encoding="utf-8"
```

```python
orgdata['seg'] = ""
orgdata.head()
```

```python
jieba.load_userdict('userdict1.txt')
```

```python
jieba.analyse.set_stop_words('stopwords1.txt')
index = 0
for documents in orgdata['low'].astype(str):
    seg_list = jieba.analyse.extract_tags(documents, 150)
    #seg_list = jieba.lcut(documents)
    orgdata['seg'].iloc[index] = seg_list
    index += 1
```

```python
orgdata['seg']
```

```python
orgdata
```

```python
subdata = orgdata[['time','fact','seg']]
```

```python
output_series = subdata.groupby(['fact'])['seg'].apply(lambda x: ','.join('%s' %a for a in orgdata['seg'])).reset_index()
```

```python
output_series
```

```python
output_series['seg'] = output_series['seg'].str.replace("', '", " ")
output_series
```

```python
from sklearn.feature_extraction.text import CountVectorizer

docs = output_series['seg'].tolist()
vec = CountVectorizer()
X = vec.fit_transform(docs)
df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
df
```

```python
coocc = df.T.dot(df)
coocc
```

```python
import plotly.express as px
fig = px.imshow(coocc)
fig.layout.height = 1000
fig.layout.width = 1000
fig.show()
```

```python
import numpy as np

filter_num_up = 3400
filter_num_down = 3599
getid = coocc.index
important = pd.DataFrame(columns = getid).T
important['relation'] = ""
dropid = []

for i in range(len(coocc)):
    result = coocc[(coocc.iloc[i] > filter_num_up) & (coocc.iloc[i] < filter_num_down)].index.tolist()
    if(len(result) > 0):
        important['relation'].loc[getid[i]] = coocc[(coocc.iloc[i] > filter_num_up) & (coocc.iloc[i] < filter_num_down)].index.tolist()
        print(important['relation'].loc[getid[i]])
    else:
        dropid.append(i)
        important = important.drop([getid[i]])    
```
